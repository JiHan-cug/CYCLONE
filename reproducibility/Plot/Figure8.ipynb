```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_method = 3 
n_sets = 5 
methodLabels = ['CYCLONE with AE', 'CYCLONE with no update',  'CYCLONE']
labels = ['MCA', 'Bct-s', 'Human Pancreas', 'Human PBMC', 'Human Lung Cell'] ## xticklabels

x = np.arange(n_sets)
total_width = 0.8
width = total_width / n_method

x = x - (total_width - width) / 2
```

```python
# load data
arivisual = [[0.6677, 0.7074, 0.7839],
             [0.8817, 0.96, 0.97],
             [0.7472, 0.7113, 0.8558],
             [0.6548, 0.81, 0.82],
             [0.56, 0.55, 0.58]]
nmivisual = [[0.77, 0.75, 0.8025],
             [0.8577, 0.9, 0.94],
             [0.82, 0.829, 0.8623],
             [0.78, 0.8189, 0.82],
             [0.71, 0.70, 0.7]]

baswvisual = [[0.96, 0.977, 0.9767],
              [0.99, 0.99, 0.9974],
              [0.8969, 0.978, 0.98],
              [0.99, 0.99, 0.9965],
              [0.897, 0.9429, 0.95]]

arivisual = np.array(arivisual)
nmivisual = np.array(nmivisual)
baswvisual = np.array(baswvisual)
```

```python
colors = ["#456990", "#48C0AA", "#EF767A"]
```

# 8A

```python
fig = plt.figure(figsize=(12, 8))

for i in range(n_method):
 plt.bar(x + i * width, arivisual[:, i], width=width, label=methodLabels[i], color=colors[i], edgecolor="white")
ax = plt.subplot()
plt.xticks(range(n_sets), labels)
plt.ylabel("ARI", family='Arial', size=20)
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])

ax.set_xticklabels(labels, family='Arial', fontsize=14)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.ylim(0.5, None)
```
