# Figure 1A

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_method = 9 # NOTE: change methods number here
n_sets = 6 # NOTE: change datasets number here
methodLabels = ['BBKNN', 'INSCT', 'Scanorama',  'Cell Blast',  'Seurat', 'Harmony', 'scVI', 'scDML', 'CYCLONE']
labels = [
 'Simulation', 'MCA', 'Bct-s', 'Human Pancreas', 'Human PBMC', 'Human Lung Cell'] ## xticklabels

x = np.arange(n_sets)
total_width = 0.8
width = total_width / n_method
x = x - (total_width - width) / 2
```

```python
# prepare data
arivisual = [[0.0301, 0.0587, 0.6578, 0.8812, 0.9286, 1, 1, 1, 1],
             [0.4047, 0.4622, 0.51421, 0.7076, 0.4245, 0.3903, 0.69835, 0.75, 0.7839],
             [0.3075, 0.97, 0.1909, 0.98, 0.3028, 0.6679, 0.97, 0.953, 0.97],
             [0.3178, 0.4338, 0.023, 0.5253, 0.436, 0.4946, 0.7254, 0.8495, 0.8558],
             [0.3742, 0.5226, 0.53778, 0.7296, 0.4321, 0.4518, 0.79515, 0.6645, 0.81459],
             [0.3075, 0.3708, 0.10745, 0.4472, 0.198, 0.3855, 0.5697, 0.311, 0.5748]]
arivisual = np.array(arivisual)
```

```python
# set colors here
colors = ["#3F626D", "#6E8989", "#AB9062", "#ECB1AE", "#DEC9C7", "#AFCDE5", "#C8C6B5", "#EFCEA0", "#ab372f"]
```

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot()

for i in range(n_method):
 plt.bar(x + i * width, arivisual[:, i], width=width, label=methodLabels[i], color=colors[i], edgecolor="white")

box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])

ax.set_xticklabels(labels, family='Arial', fontsize=14, rotation=15)
# Put a legend below current axis
ax.legend(loc='upper right', bbox_to_anchor=(1.12, 0.7), fancybox=True, shadow=False, ncol=1)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xticks(range(n_sets), labels)
plt.ylabel("ARI", family='Arial', size=20)
```
